Part2. 設計一個Function 並執行使用

學生選的是河內塔，會顯示出移動的過程與總次數。

```
import math

def hanoi(n, A, B, C):
    if n == 1:
        print(str(n) + "號圈：" + A + "→" + C)
    else:
        hanoi(n-1, A, C, B)
        print(str(n) + "號圈：" + A + "→" + C)
        hanoi(n-1, B, A, C)
        
def hanoi_amount(c):
  if c == 1:
    return 1
  else:
    return 2 ** (c-1) + hanoi_amount(c-1)
      
i = int(input(print("請輸入圈圈數量：")))

total = hanoi_amount(i)
hanoi(i, "A", "B", "C")

print("總移動次數為：" + str(total) + "次")
```
