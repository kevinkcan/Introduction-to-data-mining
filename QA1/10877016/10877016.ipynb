1. Compound Data Types
用隨機變數轉換成array,list,diction

```python
import numpy as np
a=np.random.randint(0,101,size=30) #從0到100間隨機取出30個數字
alist=[]                           #建立一個空list
print(a)                           #印出隨機取出的數字(a)
print(type(a))                     #印出其類型
alist.append(a)                    #將array(a)匯入到alist內
print(alist)                       #印出alist
print(type(alist))                 #印出其類型
dict={'random number':alist}       #建立一個dictionary(dict)
print(dict)                        #印出dict
print(type(dict))                  #印出dict類型


```

2. User-Defined Functions
將五個數值平均

```python
def five_set_average(a,b,c,d,e):
    aver=(a+b+c+d+e)/5
    return(aver)
    
print(five_set_average(1,2,3,4,5))
print(five_set_average(2,4,6,8,10))    

```
